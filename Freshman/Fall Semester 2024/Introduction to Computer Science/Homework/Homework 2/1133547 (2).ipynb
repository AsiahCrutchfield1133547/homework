```python
# Asiah Crutchfield (1133547) | 28 Nov 2024
# Using a selection sort algorithim to sort an array and then transform all negative numbers into 0's
import time


def selection(array):
  print(f"Original array: ",array)
  for i in range(0, len(array)): # iterating over the list
    minimum = i
    for j in range(i+1, len(array)-1): # iterating over the elements in the list
      if array[j] < array[minimum]:
        array[minimum], array[j] = array[j], array[minimum] # swapping elements
        minimum = j # replacing the previously swapped element with the new one

  print('New array: [', end='', flush=True) # printing the array after it is sorted
  for i in range(len(array)):
    time.sleep(0.3)
    print(array[i], end='', flush=True)
    if i != len(array)-1:
      print(end=', ', flush=True)
  print(']')

  time.sleep(0.5)

  for i in range(len(array)):
    if array[i] < 0:
      array[i] = 0

  print(f"Final array: ",array) # printing the array after the negatives are replaced with 0's

selection([4, -3, 7, -1, 9])
```
