```python
#Asiah Crutchfield (1133547) | Submission: 28 Nov 2024
#Sorting a list using the insertion sort algorithim
import time

def insert(array):
  print(f"Original array: ",array)
  for i in range(1, len(array)):
    temp = array[i] # Stores current array element
    j = i-1 # Accesses array element to the left
    while j >= 0 and array[j] > temp: # Searches for element to right as long as it is less than current element
      array[j + 1] = array[j]
      j -= 1

    array[j + 1] = temp

  print("Sorted array: [", end="", flush=True)
  for i in range(len(array)):
    print(array[i], end='', flush=True)
    if i != len(array) - 1:
      print(", ", end='', flush=True)
      time.sleep(0.5)
  print("]")

insert([8, -3, 12, -7, 5])
```
