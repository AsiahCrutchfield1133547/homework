```python
# Asiah Crutchfield (1133547) | Submitted: 28 Nov 2024
# A program that uses the bubble sort algorithim to turn a order a list from highest to lowest

import time

# Bubble sort function
def bubbleSort(list):
  print(f"Original list: ", list)
  for i in range(len(list)): # iterate over entire array
    for j in range(len(list)-i-1): # iterate over each element
      if list[j] < list[j+1]:
        list[j+1], list[j] = list[j], list[j+1]

  print("Sorted list: [", end='', flush=True)
  for i in range(len(list)):
    print(list[i], end='')
    if i != len(list) - 1:  # If it's not the last element, print a comma and space
      print(", ", end='', flush=True)
      time.sleep(0.5)  # Simulate typing effect
  print("]", end='', flush=True)

myList = [12, 7, 19, 3, 5]

bubbleSort(myList)


```
