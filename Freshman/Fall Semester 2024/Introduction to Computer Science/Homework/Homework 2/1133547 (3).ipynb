```python
# Asiah Crutchfield (1133547) | 28 Nov 2024
# Using a merge sort algorithim to sort an array and then move all even numbers to the front while keeping the order

def merge(arr):
  if len(arr) > 1:
    left = arr[:len(arr)//2] # getting indexes for the left side of the array
    right = arr[len(arr)//2:] # getting indexes for the right side of the array

    merge(left)
    merge(right)

    i = 0 # index for the left array
    j = 0 # index for the right array
    k = 0 # index for merged array

    while i < len(left) and j < len(right):
      if left[i] < right[j]:
        arr[k] = left[i]
        i += 1
      else:
        arr[k] = right[j]
        j += 1
      k += 1

    while i < len(left):
      arr[k] = left[i]
      i += 1
      k += 1

    while j < len(right):
      arr[k] = right[j]
      j += 1
      k += 1

# a function that takes the array and places the even numbers in front sequentially
def evens(num):
  evens = [even for even in num if even % 2 == 0]
  odds = [odd for odd in num if odd % 2 != 0]

  num[:] = evens + odds

array = [3, 4, 7, 2, 6, 5, 8, 1]

print(f"Original array: ", array)

merge(array)

print(f"New array: ", array)

evens(array)

print(f"Final array: ", array)
```
